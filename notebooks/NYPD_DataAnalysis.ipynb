```python
import pandas as pd

nypd_data = pd.read_csv(f'../dataset/Motor_Vehicle_Collisions_-_Crashes.csv')
```

```python
nypd_data
```

```python
nypd_data['CRASH DATE'] = pd.to_datetime(nypd_data['CRASH DATE'])
```

```python
nypd_data_2018 = nypd_data[nypd_data['CRASH DATE'] >= '2018-01-01']
nypd_data_2018 = nypd_data_2018[nypd_data_2018['CRASH DATE'] <= '2018-12-31']
```

```python
nypd_data_2018
```

```python
import folium
from matplotlib.colors import Normalize, rgb2hex
import matplotlib.cm as cm


accident_location = nypd_data_2018[[
    'LATITUDE',
    'LONGITUDE',
]]

accident_location = accident_location.round(3)

accident_location = accident_location[accident_location['LATITUDE'] <= 41.13]
accident_location = accident_location[accident_location['LATITUDE'] >= 40.12]
accident_location = accident_location[accident_location['LONGITUDE'] >= -74.40]
accident_location = accident_location[accident_location['LONGITUDE'] <= -73.19]

count_accident_locations = pd.DataFrame({
    'incident_count' : accident_location.groupby(['LATITUDE','LONGITUDE']).size()
}).reset_index()

print(count_accident_locations)

incident_count_sum = count_accident_locations['incident_count'].sum()

norm = Normalize(
    count_accident_locations['incident_count'].min(), 
    count_accident_locations['incident_count'].max()
)
tileset = r'http://{s}.basemaps.cartocdn.com/dark_all/{z}/{x}/{y}.png'

incident_map = folium.Map(
    location=[40.73, -73.93], 
    width=1024, 
    height=786,
    tiles=tileset,
    attr='© OpenStreetMap contributors, © CartoDB',
    zoom_start=12
)

for index, row in count_accident_locations.iterrows():
    if index % 100000 == 0:
        print(index)
    color=rgb2hex(cm.YlOrRd(norm(row['incident_count'])))
    folium.CircleMarker(
        location=[
            row['LATITUDE'],
            row['LONGITUDE']
        ], 
        popup=row['incident_count'],
        radius=row['incident_count'] * 2000 / incident_count_sum,
        fill_color=color, 
        line_color=color
    ).add_to(incident_map)

incident_map.save(outfile='accident_map.html')
```

```python
vehicle_types = []
for code in range(1,6):
    vehicle_code_column = f'VEHICLE TYPE CODE {code}'
    vehicle_types += list(nypd_data[vehicle_code_column].str.lower().unique())

list(set(vehicle_types))
```

```python
bike_codes = ['BICYCLE', 'E-BIK', 'Bike', 'E BIK', 'e BIK', 'E/BIK', 'BICYC', 'e-bik', 'EBIKE', 'Bicyc', 'e bik']
bike_codes = list(set([code.lower() for code in bike_codes]))
bike_codes
```

```python
bike_accidents = None
for code in range(1,6):
    vehicle_code_column = f'VEHICLE TYPE CODE {code}'
    if bike_accidents is None:
        bike_accidents = nypd_data[nypd_data[vehicle_code_column].str.lower().isin(bike_codes)]
    else:
        new_bike_accidents = nypd_data[nypd_data[vehicle_code_column].str.lower().isin(bike_codes)]
        bike_accidents = pd.concat([bike_accidents,new_bike_accidents]).drop_duplicates() 

bike_accidents
```

```python
accident_location = bike_accidents[[
    'LATITUDE',
    'LONGITUDE',
]]

accident_location = accident_location.round(3)

accident_location = accident_location[accident_location['LATITUDE'] <= 41.13]
accident_location = accident_location[accident_location['LATITUDE'] >= 40.12]
accident_location = accident_location[accident_location['LONGITUDE'] >= -74.40]
accident_location = accident_location[accident_location['LONGITUDE'] <= -73.19]

count_accident_locations = pd.DataFrame({
    'incident_count' : accident_location.groupby(['LATITUDE','LONGITUDE']).size()
}).reset_index()

print(count_accident_locations)

incident_count_sum = count_accident_locations['incident_count'].sum()

norm = Normalize(
    count_accident_locations['incident_count'].min(), 
    count_accident_locations['incident_count'].max()
)
tileset = r'http://{s}.basemaps.cartocdn.com/dark_all/{z}/{x}/{y}.png'

incident_map = folium.Map(
    location=[40.73, -73.93], 
    width=1024, 
    height=786,
    tiles=tileset,
    attr='© OpenStreetMap contributors, © CartoDB',
    zoom_start=12
)

for index, row in count_accident_locations.iterrows():
    if index % 100000 == 0:
        print(index)
    color=rgb2hex(cm.YlOrRd(norm(row['incident_count'])))
    folium.CircleMarker(
        location=[
            row['LATITUDE'],
            row['LONGITUDE']
        ], 
        popup=row['incident_count'],
        radius=row['incident_count'],
        fill_color=color, 
        line_color=color
    ).add_to(incident_map)

incident_map.save(outfile='bike_accident_map.html')
```
